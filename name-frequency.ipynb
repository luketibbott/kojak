```python
df = pd.read_csv('./data/aac_intakes.csv')
```

```python
df_name = df[['animal_id', 'name']]
```

```python
from collections import Counter
```

```python
name_freqs = Counter(df_name['name'].str.replace(r'(\*)', '').dropna())
```

```python
name_freqs
```

```python
def name_frequency(s):
    if s in name_freqs:
        return name_freqs[s]
    return 0
```

```python
df_name['name_freq'] = df_name['name'].str.replace(r'(\*)', '').apply(lambda x: name_frequency(x))
```
